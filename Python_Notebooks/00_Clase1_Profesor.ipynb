## Leyendo datos a la antigua

```python
rows = []
for i in range(len(datos)):
    rows.append(datos[i].split("\t"))
    print(i)
    
print("Hello word")

```

```python
### Creamos el datafram usando las filas que tenemos guardadas en "rows"
df = pd.DataFrame(rows[1:], columns = rows[0])

### el metodo .drop() me sirve para eliminar filas o columnas de un dataframe
### Si voy a eliminar una fila al metodo .drop() le paso el nombre de la fila (en este caso 193)
### le especifico en que axis va a hacer la eliminacion 0 para las filas y 1 para las columnas
### inplace = True me sirve para que una vez haga la eliminación actualice y sobre escriba el dataframe

df.drop(193, axis = 0, inplace = True)
```

## Dos formas equivalentes de usar el open

```python
file = open("Orofaringe.txt", "r") ## Usamos la funcion open en formato "r" de read (lectura)
datos = file.read() ### Una vez abierto el archivo para acceder a su contenido lo debo leer con el metodo .read()
file.close() ### No olvien cerrar lo que alguna vez abrieron
```

```python
#### Todo en uno de una forma elegante

with open("Orofaringe.txt", "r") as file: ### Estas dos lineas de codigo hacen exactamente lo mismo que la celda anterior
    datos2 = file.read()
    
datos2 = datos2.split("\n")

rows = []
for i in range(len(datos2)):
    rows.append(datos2[i].split("\t"))
    
df = pd.DataFrame(rows[1:], columns = rows[0])
df.drop(193, axis = 0, inplace = True)
df
```

## Optimizando un poco la lectura con open

```python
with open("Orofaringe.txt", "r") as file: ### Estas dos lineas de codigo hacen exactamente lo mismo que la celda anterior
    rows = [row.replace("\n", "").split("\t") for row in file.readlines()]
    
data = pd.DataFrame(rows[1:], columns=rows[0], dtype = float)
data
```

## Datset IRIS

```python
import json

with open("iris.json", "r") as archivo:
    datos = archivo.read()
    
df = pd.DataFrame(json.loads(datos))
```

```python
#### Y como escribo archivos con el open
with open("iris2.csv", "w") as file:
    file.write(df.to_string(index = False))
```

```python
pd.plotting.scatter_matrix(df)
```

```python
df.drop("species", axis = 1).corr() ## El metodo .corr() calcula las correlaciones de las variable numericas dentro del df
```

```python
df.describe()
```

```python
import seaborn as sns

sns.pairplot(df, hue = "species")
```

```python
from scipy import stats

stats.median_abs_deviation(df.drop("species", axis = 1))
```

```python
def miFuncion(x):
    resultado = stats.median_abs_deviation(x)
    return resultado

## Cuando el calculo que deso hacer para el agregamiento no existe como metodo dentro de pandas
## Nosotros podemos usar cualquier funcion progrmada por nosotros usando el metodo .apply() y agregando una lambda function
df.groupby("species").apply(lambda x: miFuncion(x))
```

```python
### Haciendo agregamientos de datos Para calcular la MAD sando el paquete scipy.stats
temp = df.groupby("species").apply(lambda x: stats.median_abs_deviation(x))

pd.DataFrame(list(temp), index = temp.index, columns = df.columns[:4])
```
