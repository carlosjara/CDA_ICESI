# Leyendo Datos a la antigua

```python
#Abrir archivos desconocidos - Leer cualquier tipo de archvios sin preocuparse por la extension
file = open ("Datos/orofaringe.txt","r") ## usamos la funcion open en formato r de lectura (Read) - w write (escritura) -a  append (Escribir al final) 
datos = file.read() ## Una vez abierto el archivo, lo debo leer con el metodo .read()
file.close() ## NO OLVIDAR CERRAR LOS ARCHIVOS
```

```python
file = open ("Datos/orofaringe.txt","r")
print(file.read())
file.close() ## NO OLVIDAR CERRAR LOS ARCHIVOS
```

```python
datos
```

```python
datos = datos.split("\n")
```

```python
datos[0].split("\t")
```

```python
rows = []
for i in range(len(datos)):
    rows.append(datos[i].split("\t"))
```

```python
## funcion (funciona independientemente de que un elemento exista o no)
## Creamos el data frame usando las filas que tenemos guardadas en rows
df= pd.DataFrame(rows[1:], columns = rows[0])

## Metodo funcion a elemetos que ya existen
### el metodo .drop() me sirve para eliminar filas o columnas de un dataframe
## si voy a eliminar una fila a drop, le paso el nombre de la fila (indices de rows)
## Le especifico el axis ( 1 columnas o 0 filas )
## inplace = True me sirve para que una vez haga la eliminacion , actualice y sobre escriba el dataframe
df.drop(193,axis = 0, inplace=True)
df
```

```python
with open("Datos/orofaringe.txt","r") as file:
    datos2 = file.read()
    
datos2 = datos2.split("\n")

rows = []
for i in range(len(datos2)):
    rows.append(datos2[i].split("\t"))

df= pd.DataFrame(rows[1:], columns = rows[0])
df.drop(193,axis = 0, inplace=True)    
```

```python
## haciendo agregamientos de datos
df.groupby("species").apply(lambda x: np.percentile(x, [25,50,75])) #aplica a todas las columnas 
                                                                    ##la funcion que yo le diga (lambda <input<:<output>)
# declarar funciones implicitas (lambda) 
## Cuando el calculo que deseo hacer para el agregamiento , no existe como metodo dentro 
#se puede usar, cualquier funcion programada por nosotros usando el metodo .apply() y agregando una lambda function
```

```python
def miFuncion(x): #funciones por extensibilidad 
    return stats.median_abs_deviation(x)
#ejemplo de funcion
```

```python
df.groupby("species").apply(lambda x: miFuncion(x)) #aplica a todas las columnas          
```
